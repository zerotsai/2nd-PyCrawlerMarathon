### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text,"html5lib")
print(soup)

```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find_all('em')[-1].text+"    ", d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_list = []

for i in soup.find(class_="part_list_2").find_all("h3"):
    news_list.append([i.find('em').text,i.find(class_='date').text,i.find_all('a')[-1].text])

print(news_list)
```

```python
import pandas as pd
#df = pd.DataFrame(news_list,columns=['類別','日期','標題'])
#df
columns = ['類別','日期','標題']
table = pd.DataFrame(news_list, columns=columns)
table
```

### ③ 哪一個類別的文章最多

```python
group = table.groupby('類別')
table2 = pd.DataFrame(group.size(), columns=['文章數量'])
table2.sort_values(by='文章數量', ascending=False)
```
